```python
score=0
for i in range(53):
    X= data.loc[:, data.columns.drop(column_groups[i])]
    y=data[column_groups[i][0]]
    X_train, X_test,y_train, y_test = train_test_split(X,y ,random_state=104, test_size=0.2, shuffle=True)

    from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel, Matern
    kernel = DotProduct() + WhiteKernel()+ Matern()

    from sklearn.gaussian_process import GaussianProcessRegressor
    regressor = GaussianProcessRegressor(kernel=kernel)

    regressor.fit(X_train, y_train)

    temp=regressor.score(X_test,y_test)
    score+=temp
    print(i+1," ",temp)
print(score/53)
```
